```python
# Import the imdb package.
import imdb

from os import listdir
from os.path import isfile, join

PATH = '/Users/guillaume_j/Movies/Film/'

movie_titles = [f for f in listdir(PATH)]

# Create the object that will be used to access the IMDb's database.
ia = imdb.IMDb() 

movies = []

for title in movie_titles:
    print('Raw movie title:', title)
    idx = title.find("-")
    title = title[:idx]
    if title != -1:
        s_result = ia.search_movie(title)
        if s_result:
            movie = s_result[0]
            print('Movie:', title)
            print('Title:', movie['title'])
            print('Date:', movie['year'])
            print('Genre:', movie['kind'])
            ia.update(movie)
            if 'director' in movie:
                print('Director:', movie['director'])
            movies.append(movie)
```

```python
import json 

db = {}
db['movies'] = []
for movie in movies:
    entry = {}
    entry['title'] = movie['title']
    entry['year'] = movie['year']
    entry['genre'] = movie['kind']
    if 'cover url' in movie:
        entry['cover'] = movie['cover url']
    if 'director' in movie:
        entry['director'] = movie['director'][0]['name']
    db['movies'].append(entry)

print(db['movies'][10])

with open('movie_db.json', 'w') as fp:
    json.dump(db, fp, indent=4, ensure_ascii=False)
```
